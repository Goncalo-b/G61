```python
import pandas as pd
```

```python
file = pd.read_csv("g61_Elearning_Certificates1.csv", sep=";", parse_dates=["issue_date","online_date","student_dob"], dayfirst=True).dropna(how="all")
file=file.dropna(axis=1, how="all")



```

```python
file.tail(20)
```

```python
file["platform_id"] = pd.to_numeric(file["platform_id"], errors="coerce").astype("Int64")
file["course_id"] = pd.to_numeric(file["course_id"], errors="coerce").astype("Int64")
file["certificate_id"] = pd.to_numeric(file["certificate_id"], errors="coerce").astype("Int64")
file["student_id"] = pd.to_numeric(file["student_id"], errors="coerce").astype("Int64")

```
